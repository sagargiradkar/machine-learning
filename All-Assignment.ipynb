```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
df
```

```python
x = df.iloc[:,3:]
```

```python
x
```

```python
plt.title("Unclustered Data")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.scatter(x['Annual Income (k$)'],x['Spending Score (1-100)'])
```

```python
from sklearn.cluster import KMeans,AgglomerativeClustering
```

```python
km = KMeans(n_clusters=3)
```

```python
x.shape
```

```python
km.fit_predict(x)
```

```python
# SSE
km.inertia_
```

```python
sse = []
for k in range(1,16):
    km = KMeans(n_clusters=k)
    km.fit_predict(x)
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.title("Elbow Method")
plt.xlabel("Value of K")
plt.ylabel("SSE")
plt.grid()
plt.xticks(range(1,16))
plt.plot(range(1,16),sse,marker='.',color='r')
```

```python
from sklearn.metrics import silhouette_score
```

```python
silh = []
for k in range(2,16):
    km = KMeans(n_clusters=k)
    labels = km.fit_predict(x)
    score = silhouette_score(x,labels)
    silh.append(score)
```

```python
silh
```

```python
plt.title("Silhoutte Method")
plt.xlabel("Value of K")
plt.ylabel("Silhoul Score")
plt.grid()
plt.xticks(range(2,16))
plt.bar(range(2,16),silh,color='r')
```

```python
km = KMeans(n_clusters=5,random_state=0)
```

```python
import pandas as pd
import csv
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
dataset = []
with open('Market_Basket_Optimisation.csv') as file:
    reader = csv.reader(file,delimiter=',')
    for row in reader:
        dataset += [row]
```

```python
dataset
```

```python
len(dataset)
```

```python
te = TransactionEncoder()
x = te.fit_transform(dataset)
```

```python
x
```

```python
df = pd.DataFrame(x,columns=te.columns_)
```

```python
len(te.columns_)
```

```python
df.head()
```

```python
# 1.Find Frequent itemsets
frequ_itemset = apriori(df,min_support=0.01,use_colnames=True)
```

```python
frequ_itemset
```

```python
# Find Rule 
rules = association_rules(frequ_itemset,metric='confidence',min_threshold=0.25)
```

```python
rules
```

```python
rules = rules[['antecedents','consequents','support','confidence']]
```

```python
rules.head()
```

```python
rules[rules['antecedents'] == {'cake'}]['consequents']
```
