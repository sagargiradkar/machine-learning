### Load dataset and split into features (X) and target variable (y)


```python
X=pd.DataFrame(air_data['Temp'])
y=pd.DataFrame(air_data['Ozone'])
```

#  1: Linear Regression Model and Performance Assessment


### Split the data into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.75,test_size=0.25)
```

### Train a linear regression model

```python
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)
```

### Predict using the trained model

```python
y_pred = linear_regressor.predict(X_test)
```

### Calculate MSE, MAE, and R-Squared

#### MSE 

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

#### MAE 

```python
mae = mean_absolute_error(y_test, y_pred)
mae
```

#### R^2

```python
r2 = r2_score(y_test, y_pred)
r2
```

#  2: Visualization of Simple Regression Model

```python
plt.title("Temperature-Ozone")
plt.xlabel("Temprature")
plt.ylabel("Ozone")
plt.scatter(X_test,y_test,label='actual',color='r')
plt.plot(X_test,y_pred,label='predicted',color='g')
plt.legend()
```
