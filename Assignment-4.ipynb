# Elbow graph

```python
inertias=[]
for i in range(1,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(x)
    inertias.append(kmeans.inertia_)
    print("Inertia Values:",kmeans.inertia_)
    
plt.plot(range(1,10), inertias, marker='o')
plt.title("Elbow Method")
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
#x[fit == 0, 0] selects the x-coordinates of data points belonging to cluster 0, and c='red' sets the color of those points
#to red. The s=100 parameter sets the size of the points to 100, and label is used to specify a label for the legend.
fit=kmeans.fit_predict(x)
plt.scatter(x[fit == 0,0] , x[fit == 0,1] , s=100 , c='red',label='cluster 0')
plt.scatter(x[fit ==1,0], x[fit==1,1],s=100, c='green', label='cluster 1')
plt.scatter(x[fit ==2,0], x[fit==2,1],s=100, c='pink', label='cluster 2')
plt.scatter(x[fit ==3,0], x[fit==3,1],s=100, c='blue', label='cluster 3')
plt.scatter(x[fit ==4,0], x[fit==4,1],s=100, c='orange', label='cluster 4')

plt.scatter(cent1[:, 0] , cent1[:, 1], s=200, marker= '*' ,color='k')

plt.legend()
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#"ward": This is the linkage method used for hierarchical clustering. In this case, "ward" refers to Ward's minimum variance 
#method, which is a commonly used linkage method. 
d = linkage(x,"ward")
```

# Dendrogram

```python
dendrogram(d)
plt.title('Agglomerative Clustering Dendrogram')
plt.xlabel('x')
plt.ylabel('output')
plt.show()
```
